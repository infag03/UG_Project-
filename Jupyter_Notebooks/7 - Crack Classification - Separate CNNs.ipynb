## Brick

### Import Data

```python
# Get all the data in the directory, shuffled, and reshape images.
# Multiply every value by 1/255 (this is scaling the data on a scale between 0-1)
# Trying some additional image augmentation to train the model more robustly...
# more conditions and variability to consider in each image:
    # 1. introducing variation in image brightness.
    # 2. introducing rotation range of 90 degrees.
    # 3. introducing horizontal flipping.
    
data_brick = ImageDataGenerator(rescale=1/255, 
                                horizontal_flip=True, 
                                rotation_range=90, 
                                brightness_range=[0.5,1.5]
                               ).flow_from_directory(
    
    '/Users/alexandercheng/Desktop/code/projects/crack-detection-data/Material-Classification-Crack-Sorted/Brick',
    target_size=(256,256), #squish images down to this size
    batch_size=278, #we want batch size to be all of the images in the folder
    shuffle=True,
    seed=123)
```

### Assign Images + Labels

```python
# Split images and labels.
images_brick, labels_brick = next(data_brick)
```

### Train / Validation / Test Split

```python
# Train / test split first.
# Then split again from training data into train / validation.

X_train_brick, X_test_brick, y_train_brick, y_test_brick = train_test_split(images_brick, labels_brick, test_size=0.2, random_state=123, stratify=labels_brick)
X_train_brick, X_validation_brick, y_train_brick, y_validation_brick = train_test_split(X_train_brick, y_train_brick, test_size=0.3, random_state=123, stratify=y_train_brick)
```

```python
# Check image and label shape.
# Each image should now be reshaped into 256 pixels x 256 pixels x 3 colors (RGB).

print('TRAIN')
print('Images Shape', X_train_brick.shape)
print('Labels Shape', y_train_brick.shape)
print('VALIDATION')
print('Images Shape', X_validation_brick.shape)
print('Labels Shape', y_validation_brick.shape)
print('TEST')
print('Images Shape', X_test_brick.shape)
print('Labels Shape', y_test_brick.shape)
```

### Build CNN Model

```python
from keras.applications import InceptionResNetV2
```

```python
cnn_brick = models.Sequential() #instantiate model

cnn_base_brick = InceptionResNetV2(weights='imagenet', #None or "imagenet" - we want "imagenet" because we want the trained weights.
                       include_top=False, #allows us to specify the input shape of our images
                       input_shape=(256, 256, 3)) #since "include_top=False" we need to provide an input shape for our incoming image data.
```

```python
def build_cnn(cnn_name, cnn_base_name):
    
    cnn_name.add(cnn_base_name) #add InceptionResNetV2 pre-trained model without top

    cnn_name.add(layers.Conv2D(64, (3, 3), #64 different filters, with a 3x3 filter.
                      activation='relu', #use ReLu activation functions throughout, save sigmoid / softmax function for the end layer.
                      input_shape=(256, 256,  3), #image shape input
                      padding='SAME')) #"valid" means no padding, and "same" means with padding.

# 64 bias parameters.
# 64 * (3 * 3 * 3) weight parametrs.
# Output is 64*256*256.

    cnn_name.add(layers.MaxPooling2D((2, 2))) #(2,2) reduces dimensionality of image by half.

# Output is 64*128*128.
# 32 nodes in this layer...no padding here.
# Adding padding='SAME' here because InceptionResNetV2 downsamples the input image very aggressively...
# And we would otherwise get a negative dimension size in the next pooling step.
    cnn_name.add(layers.Conv2D(32, (3, 3), activation='relu', padding='SAME')) 

# 32 bias parameters.
# 32 * (3*3*64).
# Output is 32*128*128.

    cnn_name.add(layers.MaxPooling2D((2, 2))) 

# Output is 32*64*64.
# Flattening is converting the data into a 1-dimensional array for inputting it to the next layer. 
# We flatten the output of the convolutional layers to create a single long feature vector. 
# 32 x 64 x 64 = 131,072 features once flattened.
    cnn_name.add(layers.Flatten())

# And the flattened data is connected to the final classification, fully-connected layers.
# We simply put all the pixel data in one line and make connections with the final layer. 
# Dense fully connected layer.
    cnn_name.add(layers.Dense(32, activation='relu')) 

# Final layer does the classification.
# A material cannot be both "not cracked" and "cracked" - it is either one or the other.
# So we use a "Softmax" function for classification where there is only 1 "right answer".
    cnn_name.add(layers.Dense(2, activation='softmax')) 

#Check how many trainable weights are in the model
    print('There are', len(cnn_name.trainable_weights), 'trainable weights.')
    print()

# Check how many layers are trainable
    print('There are', len(cnn_name.layers), 'trainable layers.')
    print()

# Check which layers are trainable
    for layer in cnn_name.layers:
        print(layer.name, layer.trainable)
```

```python
build_cnn(cnn_brick, cnn_base_brick)
```

### Freeze Base CNN

```python
#freeze the base_cnn model, and double-check to make sure it's frozen.
cnn_base_brick.trainable = False
for layer in cnn_brick.layers:
    print(layer.name, layer.trainable)
```

### Create Callbacks

```python
# Create callbacks
saving_weights = keras.callbacks.ModelCheckpoint(
'weights.{epoch:02d}-{val_loss:.2f}.hdf5', #save our output as we go through.
monitor='val_loss', verbose=0, save_best_only=False, #for each epoch, give me the validation loss.
save_weights_only=False, mode='auto', period=10)

reduce_lr = keras.callbacks.ReduceLROnPlateau( #lr = learning rate - we want a big step size up front to speed up the process, but smaller steps later on to ensure we don't overshoot the minimum)
monitor='val_loss', factor=0.1, patience=20,
verbose=0, mode='auto', min_delta=0.0001, min_lr=0)

nan_problem = keras.callbacks.TerminateOnNaN()

early_stop = keras.callbacks.EarlyStopping(
monitor='val_loss', min_delta=0, patience=20,
verbose=0, mode='auto', baseline=None, restore_best_weights=False)

csv_logger = keras.callbacks.CSVLogger('training.log') #write information about my model to a file
```

### Train CNN

```python
# "adam" as optimizer combines the best properties of AdaGrad and RMSProp
cnn_brick.compile(loss='binary_crossentropy',
            optimizer="adam",
            metrics=['acc'])
```

```python
cnn_brick_history = cnn_brick.fit(X_train_brick,
               y_train_brick,
               epochs=20,
               batch_size=100,
               validation_data=(X_validation_brick, y_validation_brick),
               callbacks=[csv_logger, early_stop, nan_problem, reduce_lr, saving_weights])
```

### Evaluate Accuracy + Loss

```python
hist_cnn_brick = cnn_brick_history.history
loss_values_brick = hist_cnn_brick['loss']
val_loss_values_brick = hist_cnn_brick['val_loss']
acc_values_brick = hist_cnn_brick['acc']
val_acc_values_brick = hist_cnn_brick['val_acc']
epochs_brick = range(1, len(loss_values_brick) + 1)
```

## Analysis

```python
cms = [conf_matrix_brick, conf_matrix_concrete, conf_matrix_drywall, conf_matrix_glass, conf_matrix_tile]
```

```python
#Evaluate Recall:
recall = []
for x in cms:
    recall.append(x['Predicted Cracked'][0] / (x['Predicted Not Cracked'][0] + x['Predicted Cracked'][0]))
recall
#brick
#concrete
#drywall
#glass
#tile
```

```python
#Evaluate accuracy:
accuracy = []
for x in cms:
    accuracy.append(
        (x['Predicted Cracked'][0] + x['Predicted Not Cracked'][1]) / 
        (x['Predicted Not Cracked'].sum() + x['Predicted Cracked'].sum())
    )
accuracy
    
#brick
#concrete
#drywall
#glass
#tile
```

```python
# Evaluate Precision:
precision = []
for x in cms:
    precision.append(x['Predicted Cracked'][0] / (x['Predicted Cracked'][1] + x['Predicted Cracked'][0]))
precision
    
#brick
#concrete
#drywall
#glass
#tile
```

```python
# Evaluate F1 Score:
nums = list(range(len(precision)))
f1 = []
for x in nums:
        f1.append(2 * (precision[x] * recall[x]) / (precision[x] + recall[x]))
f1

#brick
#concrete
#drywall
#glass
#tile
```

```python
materials = ['brick', 'concrete', 'drywall', 'glass', 'tile']
```

```python
metrics_df = pd.DataFrame({'Material': materials, 
                           'Accuracy': accuracy, 
                           'Precision': precision, 
                           'Recall': recall, 
                           'F1': f1})
metrics_df
```

```python
#statistics summary
metrics_df.describe()
```

```python
metrics_df_1 = pd.melt(metrics_df, id_vars=['Material']).sort_values(['variable','value'])
metrics_df_1
```

**OBSERVATIONS**
- The average accuracy is: 72.7%
- The average precision is: 92.4%
- The average recall is: 35.1%
- The average F1 score is: 47.6%


- **We have trained 1 CNN to detect deficiencies for EACH material (total of 5).**
- Clearly, these models err to the side of caution before classifying a material as "cracked".
- It seems that the 5 CNN models are apt to classify a material as generally "not cracked" unless it sees a very obvious crack.


- **Our precision is very high** - this is good to safe architects lots of time in sorting through construction photos after a site visit.
- This means lots of time saved on making a field report!


- However, to be safe and protect the safety of building users, **we really need to raise our recall metric across all models.**
- We do not want to misclassify a truly cracked surface as "not cracked" and not report a structural element that may be damaged.
- These models should more readily classify images as "cracked" for further scrutiny.

```python
def predict_image_cnn(model, path):
    img = load_img(path, target_size=(256, 256))
    plt.imshow(img)
    img = img_to_array(img)
    img = img/255
    img = np.expand_dims(img, axis=0)
    predict = model.predict(img)
    pred_name = classes[np.argmax(predict)]
    return predict.max(), pred_name
```
