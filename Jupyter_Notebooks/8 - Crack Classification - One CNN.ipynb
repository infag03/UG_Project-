## Build Model

```python
from keras.applications import InceptionResNetV2
```

```python
cnn_all = models.Sequential() #instantiate model

cnn_base_all = InceptionResNetV2(weights='imagenet', #None or "imagenet" - we want "imagenet" because we want the trained weights.
                       include_top=False, #allows us to specify the input shape of our images
                       input_shape=(256, 256, 3)) #since "include_top=False" we need to provide an input shape for our incoming image data.
```

```python
def build_cnn(cnn_name, cnn_base_name):
    
    cnn_name.add(cnn_base_name) #add InceptionResNetV2 pre-trained model without top

    cnn_name.add(layers.Conv2D(64, (3, 3), #64 different filters, with a 3x3 filter.
                      activation='relu', #use ReLu activation functions throughout, save sigmoid / softmax function for the end layer.
                      input_shape=(256, 256,  3), #image shape input
                      padding='SAME')) #"valid" means no padding, and "same" means with padding.

# 64 bias parameters.
# 64 * (3 * 3 * 3) weight parametrs.
# Output is 64*256*256.

    cnn_name.add(layers.MaxPooling2D((2, 2))) #(2,2) reduces dimensionality of image by half.

# Output is 64*128*128.
# 32 nodes in this layer...no padding here.
# Adding padding='SAME' here because InceptionResNetV2 downsamples the input image very aggressively...
# And we would otherwise get a negative dimension size in the next pooling step.
    cnn_name.add(layers.Conv2D(32, (3, 3), activation='relu', padding='SAME')) 

# 32 bias parameters.
# 32 * (3*3*64).
# Output is 32*128*128.

    cnn_name.add(layers.MaxPooling2D((2, 2))) 

# Output is 32*64*64.
# Flattening is converting the data into a 1-dimensional array for inputting it to the next layer. 
# We flatten the output of the convolutional layers to create a single long feature vector. 
# 32 x 64 x 64 = 131,072 features once flattened.
    cnn_name.add(layers.Flatten())

# And the flattened data is connected to the final classification, fully-connected layers.
# We simply put all the pixel data in one line and make connections with the final layer. 
# Dense fully connected layer.
    cnn_name.add(layers.Dense(32, activation='relu')) 

# Final layer does the classification.
# A material cannot be both "not cracked" and "cracked" - it is either one or the other.
# So we use a "Softmax" function for classification where there is only 1 "right answer".
    cnn_name.add(layers.Dense(2, activation='softmax')) 

#Check how many trainable weights are in the model
    print('There are', len(cnn_name.trainable_weights), 'trainable weights.')
    print()

# Check how many layers are trainable
    print('There are', len(cnn_name.layers), 'trainable layers.')
    print()

# Check which layers are trainable
    for layer in cnn_name.layers:
        print(layer.name, layer.trainable)
```

```python
build_cnn(cnn_all, cnn_base_all)
```

## Freeze Base Model

```python
#freeze the base_cnn model, and double-check to make sure it's frozen.
cnn_base_all.trainable = False
for layer in cnn_all.layers:
    print(layer.name, layer.trainable)
```

```python
# Create callbacks
saving_weights = keras.callbacks.ModelCheckpoint(
'weights.{epoch:02d}-{val_loss:.2f}.hdf5', #save our output as we go through.
monitor='val_loss', verbose=0, save_best_only=False, #for each epoch, give me the validation loss.
save_weights_only=False, mode='auto', period=10)

reduce_lr = keras.callbacks.ReduceLROnPlateau( #lr = learning rate - we want a big step size up front to speed up the process, but smaller steps later on to ensure we don't overshoot the minimum)
monitor='val_loss', factor=0.1, patience=20,
verbose=0, mode='auto', min_delta=0.0001, min_lr=0)

nan_problem = keras.callbacks.TerminateOnNaN()

early_stop = keras.callbacks.EarlyStopping(
monitor='val_loss', min_delta=0, patience=20,
verbose=0, mode='auto', baseline=None, restore_best_weights=False)

csv_logger = keras.callbacks.CSVLogger('training.log') #write information about my model to a file
```

___

## Train Model

```python
# "adam" as optimizer combines the best properties of AdaGrad and RMSProp
cnn_all.compile(loss='binary_crossentropy',
            optimizer="adam",
            metrics=['acc'])
```

```python
cnn_all_history = cnn_all.fit(X_train_all,
               y_train_all,
               epochs=20,
               batch_size=100,
               validation_data=(X_validation_all, y_validation_all),
               callbacks=[csv_logger, early_stop, nan_problem, reduce_lr, saving_weights])
```

### Classification Metrics

```python
accuracy = []
precision = []
recall = []
f1 = []

accuracy.append((conf_matrix_all['Predicted Cracked'][0] + conf_matrix_all['Predicted Not Cracked'][1]) / (conf_matrix_all['Predicted Cracked'].sum() + conf_matrix_all['Predicted Not Cracked'].sum()))
precision.append(conf_matrix_all['Predicted Cracked'][0] / (conf_matrix_all['Predicted Cracked'][0] + conf_matrix_all['Predicted Not Cracked'][0]))
recall.append(conf_matrix_all['Predicted Cracked'][0] / (conf_matrix_all['Predicted Cracked'][0] + conf_matrix_all['Predicted Cracked'][1]))
f1.append(2 * ((precision[0] * recall[0]) / (precision[0] + recall[0])))
```

```python
metrics_df = pd.DataFrame({'Accuracy': accuracy,
                           'Precision': precision, 
                           'Recall': recall, 
                           'F1': f1})
metrics_df
```

## Conclusion

- In conclusion, this Convolutional Neural Network should only be used in a limited respect to classify deficient work.
- With roughly 80% accuracy (every 4 out of 5 images classified correctly) this is not reliable enough to be a completely automated tool.
- Simply put, this tool is not a full replacement for human eyes looking through thousands of construction photos to find cracking, chipping, scratches, etc...


- However, this model can certainly be used in a way to increase human efficiency in detecting deficiencies.
- We can choose an appropriate cutoff point for model prediction probabilities.
- For example, let's say 90%.
    - If the model is greater than 90% sure that there is not a crack in the material, then human eyes would not need to check it.
    - If the model is less than 90% sure that there is not a crack, then we would want an architect or engineer's eyes on the photo, to double check and make sure that something was not missed.

**Future Work**

- There are several methods to potentially improve this tool.
    - More images can be collected for the material classes identified in the dataset (brick, concrete, drywall, glass, tile).
    - More classes can be created to make the CNN more robust towards other construction materials (wood, carpet, steel, plastic, etc...)
    - The architecture of the CNN can be tweaked further - adding more layers to deepen the structure may yield better results.
    - More fine tuning and allowing weights to be adjusted deeper into the pre-trained InceptionResNetV2 model may help as well.
    - It would be useful to find a way to have the algorithm automatically "mark-up" the image, to call-out the deficient area with a red box or circle for example, so architects and engineers know where to look.
